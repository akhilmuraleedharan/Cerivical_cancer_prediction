### Import the required libraries 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


import plotly.express as px
```

### Here we have a look at the dataset by using the head function

```python
cancer_df =  pd.read_csv("cervical_cancer.csv")
```

```python
cancer_df.head()
```

### Check for the type of data in the dataset as well as the non null values

```python
cancer_df.info()
```

### Also look at the statistical features of the dataset

```python
cancer_df.describe()
```

### There are a lot if `?` values in the dataset and these can be replaced with `nan` 

```python
cancer_df =  cancer_df.replace("?",np.nan)
```

```python
cancer_df.head()
```

### Check the null values in the dataset

```python
cancer_df.isna().sum()
```

### There are 2 columns `STDs: Time since first diagnosis`  and `STDs: Time since last diagnosis` which are almost empty. We remove these columns and chack for the data again using `cancer_df.info`

```python
cancer_df =  cancer_df.drop(columns=["STDs: Time since first diagnosis","STDs: Time since last diagnosis"])
```

```python
cancer_df.info()
```

### Although there are numerical values in  the dataset, there are presented as objects. We will therefore convert these to numeric

```python
cancer_df =  cancer_df.apply(pd.to_numeric)
cancer_df.info()
```

```python
cancer_df.describe()
```

### Next step is to remove the null values. Here we use the `.mean` function to fill the missing data. The data can be removed as well if needed

```python
cancer_df =  cancer_df.fillna(cancer_df.mean())
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(cancer_df.corr(), annot=True)
```

### Two columns `STDs:AIDS` and `STDs:cervical condylomatosis` show no correlation and we check the values in the columns

```python
cancer_df["STDs:AIDS"].value_counts()
```

```python
cancer_df["STDs:cervical condylomatosis"].value_counts()
```

### We can drop these columns from the dataset as they only hold one type of value`0`

```python
cancer_df =  cancer_df.drop(columns=["STDs:AIDS","STDs:cervical condylomatosis"])
```

```python
plt.figure(figsize=(30,30))
sns.heatmap(cancer_df.corr(), annot=True)
```

### Here we plot the histogram of each of the features in the datasert

```python
plt.figure(figsize=(30,30))
cancer_df.hist(bins=10, figsize=(30,30))
plt.show()
```

### Seperate the data into the target and independent variables

```python
X = cancer_df.drop(columns=["Biopsy"])
y =  cancer_df["Biopsy"]
```

```python
X
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

### There is a wide range of values in the dataset in different columns and we will use the `StandardScaler` feature from `sklearn` to scale the data and split the data into train and test samples

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test =  train_test_split(X,y,test_size=0.25 , random_state=42)
```

### Using the XGBOOST classifier to classify the data. Train it on the train data and check the accuracy of the model onto the test data

```python
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings('ignore')
xgb = XGBClassifier(random_state=0)

cv_params = {'max_depth': [2,4,8,None], 
             'learning_rate' : [0.1,0.2,0.4],
             'n_estimators': [5,10,20]
             }  

scoring =  ["accuracy","precision","recall","f1"]

xgb_cv =  GridSearchCV(xgb, cv_params, scoring=scoring, cv =4,refit='f1')
```

```python
xgb_cv.fit(X_train,y_train)
```

```python
result_train =  xgb_cv.score(X_train,y_train)
```

```python
xgb_cv.best_params_
```

```python
y_preds =  xgb_cv.best_estimator_.predict(X_test)
```

### check for the accuracy of the model using the `classification_report` and the `confusion_matrix` from sklearn.metrics

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
print(classification_report(y_test, y_preds))
```

```python
cm = confusion_matrix(y_test, y_preds)
sns.heatmap(cm, annot=True)
```

### Next use the RandomForestClassifier to classify the data and repeat the procedure

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings('ignore')
rf =  RandomForestClassifier()
cv_params = {'max_depth': [2,4,None], 
             'max_features': [0.3,0.5],
             'max_samples': [0.3,0.5],
             'min_samples_leaf': [2,4],
             'min_samples_split': [2,4],
             'n_estimators': [200,500]
             } 

scoring =  ["accuracy","precision","recall","f1"]

rf_cv =  GridSearchCV(rf, cv_params, scoring=scoring, cv =4,refit='f1')
```

```python
rf_cv.fit(X_train,y_train)
```

```python
result_train =  rf_cv.score(X_train,y_train)
```

```python
y_preds =  rf_cv.best_estimator_.predict(X_test)
```

```python
print(classification_report(y_test, y_preds))
```

```python
cm = confusion_matrix(y_test, y_preds)
sns.heatmap(cm, annot=True)
```

### The model has good metrics on the training data but not so good scores on the testing data. This can be attributed to the model being trained on a highly imbalanced dataset. One way to circumvent this is to balance the dataset by oversampling which can be carried out. 

```python

class_0, class_1 = cancer_df.Biopsy.value_counts()
cervical_class_0 = cancer_df[cancer_df["Biopsy"]==0]
cervical_class_1 = cancer_df[cancer_df["Biopsy"]==1]

cancer_df_over = cervical_class_1.sample(class_0, replace= True)
```

```python
cancer_df =  pd.concat([cervical_class_0,cancer_df_over], axis=0)
```

```python
cancer_df["Biopsy"].value_counts()
```

```python
X = cancer_df.drop(columns=["Biopsy"], axis =1)
y =  cancer_df["Biopsy"]
```

```python
y.value_counts()
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
X_train, X_test, y_train, y_test =  train_test_split(X,y,test_size=0.25 , random_state=42)

```

```python
xgb = XGBClassifier(random_state=0)

cv_params = {'max_depth': [2,4,8,None], 
             'learning_rate' : [0.1,0.2,0.4],
             'n_estimators': [5,10,20]
             }  

scoring =  ["accuracy","precision","recall","f1"]

xgb_cv =  GridSearchCV(xgb, cv_params, scoring=scoring, cv =4,refit='f1')
```

```python
xgb_cv.fit(X_train,y_train)
```

```python
result_train =  xgb_cv.score(X_train,y_train)
```

```python
y_preds =  xgb_cv.best_estimator_.predict(X_test)
```

```python
print(classification_report(y_test, y_preds))
```

```python
cm = confusion_matrix(y_test, y_preds)
sns.heatmap(cm, annot=True)
```

### As it can be seen from the model metrics, the metrics are much better than the elarlier model which was trained on unbalanced dataset.
